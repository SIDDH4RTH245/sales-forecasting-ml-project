# Sales Forecasting Project
# Step 2: EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sales_cleaned.csv", parse_dates=["Month"], index_col="Month")

df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df)
plt.title("Monthly Car Sales Over Time")
plt.show()
```

```python
df["Rolling_Mean"] = df["Sales"].rolling(window=12).mean()

df.plot(figsize=(10,5))
plt.title("Sales with Rolling Mean")
plt.show()
```

```python
sns.histplot(df["Sales"], kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Sales"])
plt.title("Sales Trend")

plt.savefig("sales_trend.png")
plt.show()
```
