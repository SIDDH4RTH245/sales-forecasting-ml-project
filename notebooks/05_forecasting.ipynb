# Step 5: Forecasting

```python
import pandas as pd
import joblib
```

```python
df = pd.read_csv("/content/sales_features.csv")

model = joblib.load("/content/sales_model.pkl")
```

```python
latest_data = df.tail(1).drop(["Sales", "Month"], axis=1)

forecast = model.predict(latest_data)

print("Next Month Sales Prediction:", forecast[0])
```
